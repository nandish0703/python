```python
import pandas as pd
pd.set_option('display.max_rows', 500)

# Make display smaller
pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('./datasets/users.csv', sep='::',
                      engine='python', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('./datasets/ratings.csv', sep='::',
                        engine='python', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('./datasets/movies.csv', sep='::',
                       engine='python', header=None, names=mnames)
```

```python
# Standard deviation of rating grouped by title
rating_std_by_title = data.groupby('title')['rating'].std()
print(rating_std_by_title)
# Filter down to active_titles
rating_std_by_title = rating_std_by_title.loc[active_titles]
print(rating_std_by_title)
# Order Series by value in descending order
rating_std_by_title.sort_values(ascending=False)[:10]
```
